# 자바스트립트 이벤트

```python
%%html
<input type="button" value="Click Me" onclick="alert('버튼 클릭(인라인)')">
```

```python
%%html
<input type="button" value="Click Me" id="target">
<script>
  (() => {
    const div = document.getElementById("target");
    div.addEventListener("click", (event) => {
      alert("버튼 클릭(스크립트)");
    });
  })();
</script>
```
