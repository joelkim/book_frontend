# 자바스트립트 이벤트

```python
from IPython.display import HTML, display

display(HTML("""
<!DOCTYPE html>
<html>
<head>
<title>Event Bubbling Example</title>
</head>
<body>
<div id="myDiv">Click Me</div>
</body>

<script>
  (() => {
    const e = document.getElementById("myDiv");
    
    e.addEventListener("click", (event) => {
      alert("clicked", event);
    });
  })();
</script>
</html>
"""))
```
